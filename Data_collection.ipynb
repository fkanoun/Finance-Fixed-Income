## Data cleaning

The variable maturity, representing the day of maturity of the bond, is static. One idea can be to turn it into a variable representing the number of days remaining before maturity which can be influencial in the fluctuation of the bond's price.

```python
for file in os.listdir("Bonds_list"):
    if(file!='.DS_Store'):
        print(file)
        df = pd.read_csv('Bonds_list/'+file)
        maturity_date = datetime.strptime(df.iloc[0]['MATURITY'], '%Y-%m-%d')
        for i in range(0, len(df)):
            date = datetime.strptime(df.iloc[i]['date'], '%Y-%m-%d')
            difference = maturity_date-date
            df.iloc[i,df.columns.get_loc('MATURITY')]=difference.days
        df.to_csv('Bonds_list/'+file)
```

In this part, the idea is to add moving averages as well as prices of previous days.

For corporate bonds market, technical analysis is still mitigated. There is still no obvious moving average to watch. The most common is over a moving one over 200 days. For the sake of trial we also added 30 and 60 days moving averages.

Also, for each date the prices of the bond for the 5 previous business days were added.

```python
def add_averages(dataframe,period):
    
    " This function adds features to the Dataframe. It adds the average price of the bond during the specified last period "
    
    new_column = 'AVERAGE_'+str(period)+'_DAYS'
    dataframe[new_column]=0
    for i in range(0,len(dataframe)):
        if(i<period):
            mean = np.mean(dataframe['PX_LAST'][0:i])
            dataframe.iloc[i,dataframe.columns.get_loc(new_column)] = mean
        else:
            mean = np.mean(dataframe['PX_LAST'][i-period:i])
            dataframe.iloc[i,dataframe.columns.get_loc(new_column)] = mean
            
    return dataframe[1:]

def add_previous_prices(dataframe,number_previous_days):
    
    "This function adds previous day price to the Dataframe"
    
    for i in range(1,number_previous_days+1):
        new_column = str(i)+'_'
        dataframe[new_column]=0
        for j in range(0,len(dataframe)):
            if(j>number_previous_days-1):
                dataframe.iloc[j,dataframe.columns.get_loc(new_column)] = dataframe.iloc[j-i]['PX_LAST']
                
    return dataframe[number_previous_days:]
```

```python
for file in os.listdir("Bonds_list"):
    if(file!='.DS_Store'):
        df = pd.read_csv('Bonds_list/'+file)
        new_df = add_averages(df,30)
        new_df = add_averages(new_df,60)
        new_df = add_averages(new_df,200)
        new_df = add_previous_prices(new_df,5)
        new_df.to_csv('Bonds_list_featured/'+file)
```

```python
#Joining the csv files into a single dataframe
features = ['date', 'BB_COMPOSITE', 'RTG_COMPOSITE_CHANGE', 'RTG_SP', 'RTG_SP_CHANGE',
            'RTG_FITCH', 'RTG_FITCH_CHANGE', 'RTG_MOODY', 'RTG_MOODY_CHANGE', 'TOT_DEBT_TO_TOT_EQY',
            'CUR_RATIO', 'IS_EPS', 'RETURN_ON_ASSET', 'RETURN_COM_EQY', 'EBITDA_TO_REVENUE', 'ISSUE_DT',
            'MATURITY', 'CALLABLE', 'NORMALIZED_PAYMENT_RANK', 'CPN_TYP', 'CPN_FREQ', 'CPN','INDUSTRY_SECTOR',
            'AVERAGE_30_DAYS', 'AVERAGE_60_DAYS', 'AVERAGE_200_DAYS', '1_','2_','3_','4_','5_',
            'PX_LAST']

final_df = pd.read_csv("Bonds_list_featured/AM300848@BMRK Corp.csv")
final_df = final_df[features]
final_df['ticker']='AM300848@BMRK Corp'

for file in os.listdir("Bonds_list_featured"):
    
    if(file!='Temp' and file!='.DS_Store' and file!='AM300848@BMRK Corp.csv'):
        
        new_df = pd.read_csv("Bonds_list_featured/"+file)
        new_df = new_df[features]
        new_df['ticker']=file.replace('.csv','')
        final_df = final_df.append(new_df,sort=True)

#Cleaning data errors in ratings
for i in range(0,len(final_df)):
    if (final_df.iloc[i,final_df.columns.get_loc('BB_COMPOSITE')]=='NR'):
        final_df.iloc[i,final_df.columns.get_loc('RTG_COMPOSITE_CHANGE')]=0
        if(i>0):
            if(final_df.iloc[i+1,final_df.columns.get_loc('BB_COMPOSITE')]!='NR'):
                final_df.iloc[i+1,final_df.columns.get_loc('RTG_COMPOSITE_CHANGE')]=0

    if (final_df.iloc[i,final_df.columns.get_loc('RTG_FITCH')]=='WD'):
        final_df.iloc[i,final_df.columns.get_loc('RTG_FITCH_CHANGE')]=0
        if(i>0):
            if(final_df.iloc[i+1,final_df.columns.get_loc('RTG_FITCH')]!='WD'):
                final_df.iloc[i+1,final_df.columns.get_loc('RTG_FITCH_CHANGE')]=0        

final_df = final_df[['ticker']+features]
final_df.to_csv('Bonds_list_featured/bonds_data.csv')
```

### S&P ratings

```python
final_df['RTG_SP'].unique()
```

```python
final_df['RTG_SP'] = final_df['RTG_SP'].apply(lambda x : str(x).replace(' *-',''))
final_df['RTG_SP'] = final_df['RTG_SP'].apply(lambda x : str(x).replace(' *+',''))
final_df['RTG_SP'] = final_df['RTG_SP'].apply(lambda x : str(x).replace(' *',''))
final_df['RTG_SP'].unique()
```

Let's clean the dataframe from null S&P ratings.

```python
final_df = final_df[(final_df['RTG_SP']!='nan') & (final_df['RTG_SP']!='NR')]
```

```python
ratings = final_df['RTG_SP'].unique()
ratings
```

```python
Proportions = [len(final_df[final_df['RTG_SP']==k]) for k in ratings]
tuples = [(ratings[k],Proportions[k]) for k in range(len(ratings))]
tuples = sorted(tuples,key=lambda tup: tup[1],reverse=True)
ratings_sorted = [k[0] for k in tuples]
counts = [k[1] for k in tuples]

plt.figure(figsize=(15,5))

plt.rcdefaults()
fig, ax = plt.subplots()
# Example data
y_pos = np.arange(len(ratings_sorted))
error = np.random.rand(len(ratings_sorted))

ax.barh(y_pos, counts, xerr=error, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(ratings_sorted)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Count')
ax.set_title('Distribution of S&P ratings')
plt.show()
```
